```python
import random
import scipy
import numpy as np
from numpy import pi
from scipy import stats
import matplotlib.pyplot as plt
import math
import scipy.integrate as integrate
import scipy.special as special 
```

```python
Na=128 #Number of particles
M=1000000 #Number of samples

Va=[] #A list where I save my random orientations
I1a=[] #An intermediate list where I save the cosines of the orientations
I2a=[]
Q1lista=[]
Q2lista=[]
Qlista=[]

for j in range(M):    
    for i in range(Na):
        za = random.random()
        #V.append(z)
        I1a.append(math.cos(2*pi*za)) 
        #print(I1)
        I2a.append(math.sin(2*pi*za)) 
    
    #print("")
    Q1a = (1/Na)*sum(I1a) 
    Q1lista.append(Q1a)
    #print(Q1list)
    Q2a = (1/Na)*sum(I2a) 
    Q2lista.append(Q2a)
    #print(Q2list)
    Qa = math.sqrt(math.pow(Q1a,2) + math.pow(Q2a,2)) 
    Qlista.append(Qa)
    #print(Qlist)

    I1a.clear()
    I2a.clear()
    #print(len(Qlist))
```

```python
# Exact curves

resulta=[]

rana = np.arange(0,0.35,0.00175)

for i in (rana):
    def g(x):
        return i*x*special.jv(0,i*x)*((special.jv(0,x/128))**128)
    a128=integrate.quad(lambda x: g(x), 0, +np.inf)[0]
    resulta.append(a128)
    
resultb=[]

ranb = np.arange(0,0.35,0.00175)

for i in (ranb):
    def g(x):
        return i*x*special.jv(0,i*x)*((special.jv(0,x/242))**242)
    a242=integrate.quad(lambda x: g(x), 0, +np.inf)[0]
    resultb.append(a242)
    
resultc=[]

ranc = np.arange(0,0.35,0.00175)

for i in (ranc):
    def g(x):
        return i*x*special.jv(0,i*x)*((special.jv(0,x/450))**450)
    a450=integrate.quad(lambda x: g(x), 0, +np.inf)[0]
    resultc.append(a450)
    
resultd=[]

rand = np.arange(0,0.35,0.00175)

for i in (rand):
    def g(x):
        return i*x*special.jv(0,i*x)*((special.jv(0,x/800))**800)
    a800=integrate.quad(lambda x: g(x), 0, +np.inf)[0]
    resultd.append(a800)
```

```python
# clt curves

def g128(x):
    return 2*128*x*np.exp(-128*(x**2))

def g242(x):
    return 2*242*x*np.exp(-242*(x**2))

def g450(x):
    return 2*450*x*np.exp(-450*(x**2))

def g800(x):
    return 2*800*x*np.exp(-800*(x**2))
```

```python
# ldt curves

def Jacob(x):
    return -(special.iv(1,x)*special.iv(1,x))/(special.iv(0,x)*special.iv(0,x)) + (special.iv(0,x) + special.iv(2,x))/(2*special.iv(0,x))

def Ratio(x):
    return special.iv(1,x)/special.iv(0,x)

def norm128(x):
    return Ratio(x)*Jacob(x)*np.exp(128*np.log(special.iv(0,x)))*np.exp(-128*x*Ratio(x))

def norm242(x):
    return Ratio(x)*Jacob(x)*np.exp(-242*x*Ratio(x))*((special.iv(0,x))**242)

def norm450(x):
    return Ratio(x)*Jacob(x)*np.exp(450*np.log(special.iv(0,x)))*np.exp(-450*x*Ratio(x))

def norm800(x):
    return Ratio(x)*Jacob(x)*np.exp(800*np.log(special.iv(0,x)))*np.exp(-800*x*Ratio(x))

C128=integrate.quad(lambda x: norm128(x), 0, 6)[0]
C242=integrate.quad(lambda x: norm242(x), 0, 4)[0]
C450=integrate.quad(lambda x: norm450(x), 0, 2)[0]
C800=integrate.quad(lambda x: norm800(x), 0, 1)[0]

gamma1 = np.arange(0, 15, 0.01)
m = special.iv(1,gamma1)/special.iv(0,gamma1)

Exact128 = 128*(gamma1*m - np.log(special.iv(0,gamma1)))
Exact242 = 242*(gamma1*m - np.log(special.iv(0,gamma1)))
Exact450 = 450*(gamma1*m - np.log(special.iv(0,gamma1)))
Exact800 = 800*(gamma1*m - np.log(special.iv(0,gamma1)))

prob128 = (1/C128)*m*np.exp(-Exact128)
prob242 = (1/C242)*m*np.exp(-Exact242)
prob450 = (1/C450)*m*np.exp(-Exact450)
prob800 = (1/C800)*m*np.exp(-Exact800)
```

```python
# The Pade curves

def Pade128(x):
    return x*np.exp(-128*4*x)*np.exp(128*(27/2)*np.log(1 + x/3))*np.exp(128*(1/2)*np.log(1-x))

D128=integrate.quad(lambda x: Pade128(x), 0, 1)[0]

def Pade242(x):
    return x*np.exp(-242*4*x)*np.exp(242*(27/2)*np.log(1 + x/3))*np.exp(242*(1/2)*np.log(1-x))

D242=integrate.quad(lambda x: Pade242(x), 0, 0.5)[0]

def Pade450(x):
    return x*np.exp(-450*4*x)*np.exp(450*(27/2)*np.log(1 + x/3))*np.exp(450*(1/2)*np.log(1-x))

D450=integrate.quad(lambda x: Pade450(x), 0, 0.3)[0]

def Pade800(x):
    return x*np.exp(-800*4*x)*np.exp(800*(27/2)*np.log(1 + x/3))*np.exp(800*(1/2)*np.log(1-x))

D800=integrate.quad(lambda x: Pade800(x), 0, 0.2)[0]
```
